```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/tracks_list.csv',sep=";",header=0)
df = df.drop(['lastfm_music_url','lastfm_artist_url','videoID_youtube'], axis=1)
print(df.shape)
```

```python
df = df.set_index('uuid')
df.head()
```

```python
df[df['title_music']=='Old Town Road'].reset_index().drop(['uuid','style','artist','title_music'],axis=1).to_csv('oldtown.csv',sep=";",header=True)
```

```python
df = df.drop(['artist','title_music'], axis=1)
```

```python
sns.pairplot(df, hue="style")
```

```python
sns.swarmplot(x="chroma_stft", y="mfcc", hue="style",
              palette=["r", "c"], data=df)
```
